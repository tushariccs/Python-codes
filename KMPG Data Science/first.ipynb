```python
#Let us perform some calculations for the data of the movies
#find average runtime of the data
movies['Runtime'].mean()
print(movies['Runtime'].mean())



#find IMDB median
print(movies['IMDb'].median())
```

```python
# filtering the data
# Problem statement:Find the numbers of movie in english

#Step 1: creating a filter condition
fc1 = movies['Language'] == 'English'
# if u want hindi movies then change 'English' by 'Hindi'
# print(fc1)

#Step 2 : Apply the filter condition
movies[fc1]
```

```python
#find the numbers of movies available in netflix with an IMDB rating of greater than eqaul to 9
#Step 1 :creating filter condition
fc2 = movies['Platform'] == 'Netflix'
movies[fc2]
fc3  = movies['IMDb'] >= 9 
movies[fc3]

#Step 2 : Applying filter condition
movies[fc2 & fc3]
movies[fc2 & fc3].shape

#To get the value of row and column
# movies[fc2 & fc3].shape[0]
# movies[fc2 & fc3].shape[1]
```

```python
# find the movies from language equal to hindi and having rotten greater than equal to 80
#Step 1 :creating filter condition
fc2 = movies['Language'] == 'Hindi'
movies[fc2]
fc3  = movies['Rotten Tomatoes'] >= 80
movies[fc3]

#Step 2 : Applying filter condition
movies[fc2 & fc3]

movies[fc2 & fc3].shape
movies[fc2 | fc3]
movies[fc2 | fc3].head(10)
movies[fc2 | fc3].tail(10)
```

```python
#Grouping of data
#PS:What is avearge IMDb rating score for each plateform
movies.groupby('Platform')[['IMDb']].mean()
```

```python
#PS: find the average rotten tomato score for movies under genres under each platform.
movies.groupby(['Platform','Genres'])[['Rotten Tomatoes']].mean().reset_index()
```

```python
# PS:Find the average runtime of movies for each country
movies.groupby('Country')[['Runtime']].mean().reset_index()

#Arrange the values in ascending order of Runtime
movies.groupby('Country')[['Runtime']].mean().reset_index().sort_values(by='Runtime',ascending=False)
```
